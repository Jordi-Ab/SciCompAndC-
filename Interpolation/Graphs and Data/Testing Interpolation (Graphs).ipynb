### Ok, that didn't work. The  problem seems to be at the function end points.

To fix this issue we will use Chebyshev Nodes, wich are not equally spaced, they are clustered towards the endpoints.
They are defined as:

$ \large n_{j} = \cos({\frac{j \pi}{n} }); \space\space  j = 0,...,\text{n_nodes}$

to map them over an interval, we make this:

$ \large n_{j} = \frac{b+a}{2} - \frac{b-a}{2}\cos({\frac{j \pi}{n} }); \space\space  j = 0,...,\text{n_nodes}$

Chebyshev Nodes:

```python
def chebyshevNodes(low, high, n):
    nodes = np.zeros(n+1)
    nodes[0] = low;
    for i in range(n):
        nodes[i] = ((high + low)/2) - ((high - low)/2)*np.cos(i*np.pi/n);
    
    nodes[n] = high;
    return nodes;
```

```python
print(chebyshevNodes(-5,5, 10))
pt.plot(chebyshevNodes(-5, 5, 10),np.zeros(11),'o')
```
