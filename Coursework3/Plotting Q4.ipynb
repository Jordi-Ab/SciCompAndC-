```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as pt
from mpl_toolkits.mplot3d import Axes3D
% matplotlib inline
```

```python
def ChebyshevGrid1D(n):
    assert(n>0)
    nodes = []
    for j in range(0,n+1):
        nodes.append(np.cos(j*np.pi/n));
    return np.array(nodes)
```
