```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d, Axes3D
% matplotlib inline
```

```python
def ChebyshevGrid1D(n):
    assert(n>0)
    nodes = []
    for j in range(0,n+1):
        nodes.append(np.cos(j*np.pi/n));
    return np.array(nodes)
```

```python
x = ChebyshevGrid1D(Nx)
y = ChebyshevGrid1D(Ny)
```

```python
array = np.loadtxt('OutputData/StormVerlet_Q4_output.dat')
```

```python
ts = array[:, 0]
us = array[:,1:]
```

```python
Nx = 40
Ny = 40
```

```python
plot_gap = 1/3
```

```python
def plot(t):
    # Reshape long 1D results onto 2D grid:
    uu = np.zeros((Nx+1,Ny+1));
    uu[1:-1,1:-1] = us[t].reshape((Nx-1,Ny-1)).T
    xx,yy = np.meshgrid(x,y)
    fig = plt.figure(figsize=plt.figaspect(0.5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(xx,  yy,  uu, cstride=1,rstride=1,color='w')
    ax.azim = -138; ax.elev = 25
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u')
    plt.title('t = '+str('%.5f' % t)); ax.azim = -138; ax.elev = 38; ax.set_zlim3d(-0.2,1)
    plt.show()
```

```python
# Plot at time 0
plot(0)
```

```python
# Plot at time 1/3
plot(int(len(us)*plot_gap)-1)
```

```python
# Plot at time 2/3
plot(int(len(us)*2*plot_gap)-1)
```

```python
# Plot at time 1
plot(int(len(us)*3*plot_gap)-1)
```
