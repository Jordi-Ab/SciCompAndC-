```python
import numpy as np
import matplotlib.pyplot as plt
import ODESolver as ode
from chebyshev import *
from scipy.interpolate import interp2d
from mpl_toolkits.mplot3d import axes3d, Axes3D
%matplotlib inline
```

```python
def initConditionF(x, y):
    return np.exp( -40*( (x-0.4)**2 ) - 40.0*(y*y) );
```

```python
# Initialize all the necessary info for Discretizing the problem
Nx = 40
Ny = 40
```

```python
x = ChebyshevGrid1D(Nx)
y = ChebyshevGrid1D(Ny)
```

##### Discrete Laplacian Operator

```python
# Interior second order Diff Matrix
D2_x = SecondOrderDifferentiationMatrix(x)[1:-1,1:-1]
# Interior second order Diff Matrix
D2_y = SecondOrderDifferentiationMatrix(y)[1:-1,1:-1]
Ix = np.eye(Nx-1)
Iy = np.eye(Ny-1)
L = np.kron(D2_x, Iy) + np.kron(Ix, D2_y)   # Laplacian 
```

```python
print(D2_x.shape)
print(Ix.shape)
print(L.shape)
```

```python
xx, yy = np.meshgrid(x[1:-1],y[1:-1]) # 2-D MeshGrid of Interior Points.
Nm1 = Nx-1
xx = xx.T.reshape(Nm1*Nm1) # stretch 2D grids to 1D vectors
yy = yy.T.reshape(Nm1*Nm1) # Transpose because default storage in numpy arrays is by rows.
```

#### Solving the Discretized System of Second Order PDE's:

```python
c = 1
t0 = 0
T = 1
h = 6/(Nx**2)
n = (T-t0)/h

# initial conditions for interior points.
u0 = initConditionF(xx, yy)
v0 = np.zeros(len(u0))
```

```python
def f(t,u):
    return (c*c)*(L.dot(u))
```

```python
method = ode.StormVerlet(f)
method.setInitialCondition(u0)
method.setInitialVelocity(v0)
```

```python
time_interval = np.linspace(t0, T, n)
ts, us = method.solve(time_interval)
```

```python
plot_gap = 1/3
```

```python
def plot(t):
    # Reshape long 1D results onto 2D grid:
    uu = np.zeros((Nx+1,Ny+1));
    uu[1:-1,1:-1] = us[t].reshape((Nx-1,Ny-1)).T
    xx,yy = np.meshgrid(x,y)
    fig = plt.figure(figsize=plt.figaspect(0.5))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(xx,  yy,  uu, cstride=1,rstride=1,color='w')
    #ax.plot_surface(xxx, yyy, uuu,color = 'w')
    ax.azim = -138; ax.elev = 25
    ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('u')
    plt.title('t = '+str('%.5f' % t)); ax.azim = -138; ax.elev = 38; ax.set_zlim3d(-0.2,1)
    plt.show()
```

```python
# Plot at time 0
plot(0)
```
