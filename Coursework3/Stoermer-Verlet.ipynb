```python
import numpy as np
import matplotlib.pyplot as pt
import ODESolver as ode
%matplotlib inline
```

```python
def f(t,u):
    return -(a**2)*u + np.sin(2*t)

def true_sol(t):
    c = a*((a**2)-4)
    return (c*np.cos(a*t) + a*np.sin(2*t) - 2*np.sin(a*t))/c
```

```python
# Initialize all necessary info:
a=1
T = 20
t0 = 0
h = .01
n = (T-t0)/h
y0 = 1 # Initial State
v0 = 0 # Initial Velocity
```

```python
# Instantiate Method:
method = ode.StormVerlet(f)
method.setInitialCondition(y0)
method.setInitialVelocity(v0)
```

```python
# Define time interval
time_interval = np.linspace(t0,T, n)
# Solve using the given time interval
ts, ys = method.solve(time_interval)
```

```python
# Plot true solution versus computed solution.
fig = pt.figure()
plt = fig.add_subplot(111)
pt.ylim([-2,2])
plt.plot(time_interval, true_sol(time_interval), label="true")
plt.plot(ts, ys, 'o-', label = 'Storm-Verlet, h='+str(h))
plt.legend(loc='upper left')
```
